```python
import tensorflow as tf
```

```python
finn = tf.keras.Sequential()
finn.add(tf.keras.layers.Dense(units = 2, input_shape = [1]))
```

```python
finn.compile(optimizer = tf.keras.optimizers.Adam(1), loss = 'mean_squared_error')
```

```python
d = [0, 1, 2, 3, 4, 5]
e = [6, 7, 8, 9, 10, 11]
```

```python
finn.fit(d, e, epochs = 500)
```

```python
print(finn.predict([5]))
```
